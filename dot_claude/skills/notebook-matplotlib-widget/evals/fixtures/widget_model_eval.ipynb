# Model evaluation

Loads predictions and computes basic diagnostics.

```python
%matplotlib widget
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

OUT = Path.cwd() / 'figures'
OUT.mkdir(exist_ok=True)
df = pd.read_csv('predictions.csv')  # columns: y_true, y_pred
```

```python
# Calibration scatter
fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(df.y_true, df.y_pred, alpha=0.3, s=10)
lim = max(df.y_true.max(), df.y_pred.max())
ax.plot([0, lim], [0, lim], 'k--', alpha=0.5)
ax.set_xlabel('y_true')
ax.set_ylabel('y_pred')
fig.savefig(OUT / 'calibration.png')
plt.show()
```
