# Quarterly sales analysis

Loads CSV, computes totals, makes a few plots.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

OUT = Path.cwd() / 'figures'
OUT.mkdir(exist_ok=True)
df = pd.read_csv('sales.csv')
```

```python
# Monthly totals by region
monthly = df.groupby(['region', 'month'])['revenue'].sum().unstack(0)
fig, ax = plt.subplots(figsize=(10, 6))
monthly.plot(kind='bar', ax=ax)
ax.set_ylabel('revenue')
fig.savefig(OUT / 'monthly_by_region.png')
plt.show()
plt.close(fig)
```

```python
# Top 10 products
top = df.groupby('product')['revenue'].sum().nlargest(10)
fig, ax = plt.subplots(figsize=(10, 6))
top.plot(kind='barh', ax=ax)
fig.savefig(OUT / 'top10_products.png')
plt.show(); plt.close(fig)
```
